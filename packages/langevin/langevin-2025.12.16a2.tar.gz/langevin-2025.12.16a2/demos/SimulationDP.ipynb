## Simulation

```python
sim_name: str = "a1p19050_b1_D0p04_η1_x31_y31_Δx1_Δt0p1"
# sim_name: str = "a0p28000_b1_D1p00_η1_x62_y62_Δx1_Δt0p1"
info_path: list[str] = [sim_name]
info: dict
_, info = read_info(info_path, dplvn)
_
```

```python
simulation = Simulation(
    name=sim_name, 
    path=info_path, 
    info=info, 
    do_snapshot_grid=True, 
    do_verbose=True,
)
```

```python
simulation.initialize()
```

```python
simulation.run()
```

<hr>

## Post-processing

```python
simulation.plot()
```

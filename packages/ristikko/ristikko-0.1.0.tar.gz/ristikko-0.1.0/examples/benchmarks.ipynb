```python
import numpy as np
import matplotlib.pyplot as plt
from keino import Params
from ristikko import System, Space, models
import timeit, functools
from tqdm.auto import tqdm

mem = joblib.Memory("cache")
```

```python
def test(N1, N2):
    params = Params(dict(
        N1 = N1,
        N2 = N2,
        mu = 0,
        t = 1,
        pbc = False,
    ))
    lattice = models.Triangular()
    realspace = System(Space.RealSpace, lattice, params)
    _ = realspace.lattice(params, Space.RealSpace)

N2 = 1
N1s = np.linspace(10, 5000, 33, dtype=int)
data = np.zeros((N1s.size, 10))
with tqdm(total=N1s.size) as pbar:
    for i in range(N1s.size):
        data[i] = timeit.repeat(functools.partial(test, N1s[i], N2), number=1, repeat=10)
        pbar.update(1)
```

```python
y = np.mean(data, axis=-1)
err = np.std(data, axis=-1)
plt.plot(N1s, y, ".", color="black")
plt.errorbar(N1s, y, yerr=err, color="black", capsize=5, fmt=".")
plt.xlabel("N1")
plt.ylabel("time to generate hops [s]")
```

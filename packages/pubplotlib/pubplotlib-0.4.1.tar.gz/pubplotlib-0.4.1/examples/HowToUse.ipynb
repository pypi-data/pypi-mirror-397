# How to Use PubPlotLib

Create publication-quality figures with journal-specific styles and sizes.

### Quick Start: Import and Set a Journal

Import the package and set your target journal. This will automatically apply the correct style and sizing for all your figures.


```python
import numpy as np
import matplotlib.pyplot as plt
import pubplotlib as pplt

# Set global style
pplt.style.use('aanda')
```

## Basic Plotting

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, ax = pplt.subplots(style='apj')
ax.plot(x, y, label='sin(x)')
ax.legend()
plt.show()
```

## Two-Column Figures

## Custom Aspect Ratio

```python
fig, ax = pplt.subplots(height_ratio=0.5)
ax.plot(x, y, color='C2', label='sin(x)')
ax.legend()
plt.show()
```

## Custom Sizing

```python
style = pplt.style.get()
print("One column width:", style.onecol)
print("Two column width:", style.twocol)

fig, ax = plt.subplots(figsize=(style.onecol, style.onecol * 0.5))
ax.plot(x, y, color='red', label='sin(x)')
ax.legend()
plt.show()
```

## Custom Style

```python
my_style = pplt.Style(
    name="my_style",
    onecol=10.,
    twocol=20.,
    mplstyle=None
)

fig, ax = pplt.subplots(style=my_style)
ax.plot(x, y, color='red', label='sin(x)')
ax.legend()
plt.show()
```

## Ticks and Formatters

```python
# Log plot with ticks and formatters
fig, ax = pplt.subplots(style='apj')
ax.plot(x, 10**(y*4), label='10^sin(x)')
ax.set_yscale('log')
ax.legend()
pplt.set_ticks(ax)
pplt.set_formatter(ax)
plt.show()

# Format only x-axis
fig, ax = pplt.subplots(style='apj')
ax.plot(x, 10**(y*4), label='10^sin(x)')
ax.set_yscale('log')
ax.legend()
pplt.set_ticks(ax)
pplt.set_formatter(ax, axis='x')
plt.show()
```

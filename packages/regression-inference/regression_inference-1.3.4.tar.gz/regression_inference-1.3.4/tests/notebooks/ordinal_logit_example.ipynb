```python
import pandas as pd
from regression_inference import OrdinalLogisticRegression, summary

import statsmodels.api as sm
anes_data = sm.datasets.anes96.load()
X, y =pd.DataFrame(anes_data.exog), pd.Series(anes_data.endog)
```

```python
model = OrdinalLogisticRegression().fit(X, y, alpha=0.05, tol=1e-8, max_iter=250, adj_cutpoints=False, cuda=True)
print(model)
```

```python
model_adj = OrdinalLogisticRegression().fit(X, y, alpha=0.05, tol=1e-8, max_iter=250, adj_cutpoints=True, cuda=False)
print(model_adj)
```

```python
pd.DataFrame(model.variance_inflation_factor())
```

```python
pd.DataFrame(model.inference_table())
```

```python
pd.DataFrame(model_adj.inference_table())
```

```python
pd.DataFrame(model.predict([1, 1, 32, 4, 90], return_table=False))
```

```python
predictions = pd.DataFrame(model.predict([1, 1, 32, 4, 90], return_table=True))
predictions
```

```python
pd.DataFrame(model_adj.predict([1, 1, 32, 4, 90], return_table=False))
```

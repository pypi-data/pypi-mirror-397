```python
import pandas as pd
import numpy as np
from regression_inference import MultinomialLogisticRegression, summary
import statsmodels.api as sm
```

```python
anes_data = sm.datasets.anes96.load()
X=pd.DataFrame(sm.add_constant(anes_data.exog))
y=pd.Series(anes_data.endog)
```

```python
model = MultinomialLogisticRegression().fit(X, y, cuda=True, cov_type="HC0")
print(model)
```

```python
model = MultinomialLogisticRegression().fit(X, y, cov_type="HC0")
print(model)
```

```python
pd.DataFrame(model.inference_table())
```

```python
pd.DataFrame(model.variance_inflation_factor())
```

```python
model.predict( np.array([[3.9, 4, 80, 6.0, 10]]))
```

```python
prediction_set = [
    np.array([[3.9, 4, 80, 6.0, 10]]),
]

predictions = pd.concat([pd.DataFrame(model.predict(test_set, return_table=True)) for test_set in prediction_set], ignore_index=True)

predictions
```
